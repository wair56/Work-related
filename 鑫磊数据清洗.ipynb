```python
#异常表查看
with open(file_dir+'basic_clean_异常.csv', 'r', encoding='utf-8') as f:
    df=pd.read_csv(f, dtype=str, index_col=False)
df['上报时间']=pd.to_datetime(df['上报时间'],errors='coerce')
df['受理时间']=pd.to_datetime(df['受理时间'],errors='coerce')
print (character(df))
```

```python
#异常编号处理
df=df[df['异常编号']!='理赔部']
#时间处理
df=df[df['上报时间'].notna()]
#异常件数
# df['aaa']=df['受理状态']
# df['受理状态']=df['异常件数']
# df['异常件数']=df['aaa']
# del df['aaa']
df['异常件数']=pd.to_numeric(df['异常件数'])
#关键字段转换字符串
df['运单号']=df['运单号'].map(strs)
```

```python
#异常表关闭,写回
df.to_csv('deep_basic_clean_异常.csv', index=False, encoding='utf_8_sig')
```
